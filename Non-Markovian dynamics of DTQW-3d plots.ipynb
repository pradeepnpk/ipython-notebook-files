```python
from qutip import *
import numpy as np
from cmath import pi, sin, cos, sqrt,exp; from math import radians
import matplotlib.pyplot as plt; from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns;	sns.set_style('ticks')
import time
from multiprocessing import Pool
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

#Basis states
ket0 = basis(2,0)
ket1 = basis(2,1)
psip = (basis(2,0)+basis(2,1)*1j).unit()
psim = (basis(2,0)-basis(2,1)*1j).unit()
ket00 = tensor(ket0,ket0);ket11 = tensor(ket1,ket1)
ket10 = tensor(ket1,ket0);ket01 = tensor(ket0,ket1)

#Walk operator
def Walk(coin_angle,step_size):
	P = 2*step_size+1
	coin = qutip.Qobj([[cos(radians(coin_angle)), sin(radians(coin_angle))],
				       [sin(radians(coin_angle)), -cos(radians(coin_angle))]])
	
	shift_p = qutip.Qobj(np.roll(np.eye(P), 1, axis=0))
	shift_m = qutip.Qobj(np.roll(np.eye(P),-1, axis=0))
	shift = tensor(ket0*ket0.dag(),shift_p) + tensor(ket1*ket1.dag(),shift_m)	
	
	U = shift*(tensor(coin,qeye(P)))
	return U

def Qwalk_gen(step_size,qubit_state,coin_angle,freq,amp):
    P=2*step_size+1
    P_state = basis(P,step_size)						
    Psi = ket2dm(tensor(qubit_state,P_state))		
    W = Walk(coin_angle,step_size) 

    for t in range(step_size):	
        nu = t*freq; mu = sqrt((2.0*amp*(1/freq))**2-1.0)
        gamma = exp(-nu)*(np.cos(nu*mu)+np.sin(nu*mu)/mu)
        K1 = tensor(sqrt((1+gamma)/2.0)*qeye(2), qeye(P))
        K2 = tensor(sqrt((1-gamma)/2.0)*sigmaz(),qeye(P))
        Psi = W*(K1*Psi*K1.dag()+K2*Psi*K2.dag())*W.dag()
    
    Prob=[];step=[]
    for k in range(0,P,1):
        Proj = basis(P,k)*basis(P,k).dag()
        Measure = tensor(qeye(2),Proj)
        q = abs((Psi*Measure).tr())
        Prob.append(q)
        step.append(k)
    return Prob

def Variance_rate(amp):
    variance=[0]; step_size=55
    for step in range(1,step_size):
        V=[]; freq=0.001
        qwalk = Qwalk_gen(step,psip,45,freq,amp)
        sites=2*step+1
        site = [-step+i for i in range(0,sites)]
        for i in site:
            V.append(qwalk[i]*pow(site[i]-np.average(qwalk),2))
        variance.append(sum(V))
    return variance

def Walk_variance(freq):
    v=[0]
    step_size = 11;qubit_state=psip;coin_angle=45;amp=1.0
    qw = Qwalk_gen(step_size,qubit_state,coin_angle,freq,amp);v=[]
    sites=2*step_size+1
    site = [-step_size+i for i in range(0,sites)]
    for i in site:
        v.append(qw[i]*pow(site[i]-np.average(qw),2))
    return sum(v) 

def von_entropy(state):
    ev = state.eigenenergies()
    Vn_ent=[]
    for i in range(len(ev)):
        Vn_ent.append(-ev[i]*np.log2(ev[i])) 
    Vn_ent = filter(lambda v: v==v, Vn_ent) #to remove nans from the list
    return sum(Vn_ent)

def mutual_info(rho):
    rhoA = rho.ptrace(0)
    rhoB = rho.ptrace(1)
    out =  von_entropy(rhoA)+von_entropy(rhoB)-von_entropy(rho)
    return out

def projectors(qstate):
    ev, evec = qstate.eigenstates()
    m=[]
    for l in range(len(evec)):
        m.append(ket2dm(evec[l]))
    return m

def quantum_mid(steps,freq,amp):
    proj_w=[];proj_rho=[]
    rho = Qwalk_gen(steps,psip,45,freq,amp)
    rho_c = rho.ptrace(0); rho_p = rho.ptrace(1)
    proj_c = projectors(rho_c); proj_p = projectors(rho_p)
    for n in range(len(proj_c)):
        for p in range(len(proj_p)):
            M = tensor(proj_c[n],proj_p[p])
            proj_rho.append(M*rho*M.dag())
    rhoMID = sum(proj_rho)
    Im_rho = mutual_info(rho)
    Im_mid = mutual_info(rhoMID)
    return Im_rho-Im_mid

#Quantum discord:
def quantum_discord(steps):
    rho_pc=Qwalk_gen(steps,psip,45,0.001,0.05)
    sites = 2*t+1
    Discord=[]
    for teta in range(0,180,9):
        for phi in range(0,360,13):
            E = qubit_gen(teta,phi).eigenstates()
            Eval = E[0];Ev = E[1]; Evec0=ket2dm(Ev[0]); Evec1=ket2dm(Ev[1])
            
            Ipc = von_entropy(rho_pc.ptrace(0))-von_entropy(rho_pc)
            p0=(tensor(Evec0,qeye(sites))*rho_pc).tr(); p1=(tensor(Evec1,qeye(sites))*rho_pc).tr()
            
            rho_c1 = von_entropy((tensor(Evec0,qeye(sites))*rho_pc*tensor(Evec0,qeye(sites)))/p0)
            rho_c2 = von_entropy((tensor(Evec1,qeye(sites))*rho_pc*tensor(Evec1,qeye(sites)))/p1)
            z = Ipc+(p0*rho_c1+p1*rho_c2)
            Discord.append(z.real)
    return min(Discord)

#my func trace distance
def TraceD(rho1,rho2):
    diff = rho1 - rho2
    Eval = diff.eigenenergies()
    return 0.5*sum(abs(Eval))

#Bloch sphere sampling
def Qubit_gen(steps,phi):
    psi = cos(radians(steps))*ket0+complex(cos(radians(phi)),sin(radians(phi)))*sin(radians(steps))*ket1
    return ket2dm(psi)   

def walk_bellstates(steps,qubit_state,coin_angle,freq,amp):
    sites=2*steps+1
    Rho_r = Qwalk_gen(steps,qubit_state,coin_angle,freq,amp).ptrace(1)
    
    diag = Rho_r.diag()
    l = len(diag)                                      #length of the list
    pivot = len(diag)/2                                #find mid point
    right = diag[pivot]+sum([diag[p1] for p1 in range(0,pivot-1)]) #sum over left
    left = sum([diag[p2] for p2 in range(pivot+2,len(diag))]) #sum over right
    
    alpha = right+left                                 #probabliity of |00><00|
    beta = diag[pivot+1]                               #probablity of +-1 lattice positions
    
    dummy = Rho_r[((sites-2)/2)]                       #<+1|-1> appears right next to the row corresponding to the +1 or -1
    temp = dummy[0]                                    #Usually one row above the center row (check notes for more)
    num = abs(temp[(len(temp)/2+1)])
    gam = complex(0,num)
    gamc = gam.conjugate()
    
    reduced = (alpha*ket00*ket00.dag()+beta*ket10*ket10.dag()+beta*ket01*ket01.dag()
                +gam*(ket10*ket01.dag())+gamc*(ket01*ket10.dag()))
    return reduced

def quantum_concurrance(rho_in):
    rho_til = tensor(sigmay(),sigmay())*(rho_in.conj())*tensor(sigmay(),sigmay())
    rho_mod = (rho_in.sqrtm()*rho_til*rho_in.sqrtm()).sqrtm()
    evals = rho_mod.eigenenergies()
    evals = sorted(abs(evals))
    esum = evals[3]-evals[2]-evals[1]-evals[0]
    return max(0,esum)
```
