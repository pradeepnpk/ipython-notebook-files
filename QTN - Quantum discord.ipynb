```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt
from cmath import pi, sin, cos, sqrt,exp; from math import radians
import seaborn as sns;	sns.set_style('ticks')
import time
from multiprocessing import Pool
import random as rand
import time as time
from scipy import linalg

#Basis states:
ket0 = basis(2,0).unit()
ket1 = basis(2,1).unit()
psip = (basis(2,0)+basis(2,1)*1j).unit()
psim = (basis(2,0)-basis(2,1)*1j).unit()

#Walk operator:
def Walk(theta,step_size):
	P = 2*step_size+1
	coin = qutip.Qobj([[cos(radians(theta)), sin(radians(theta))],
				       [sin(radians(theta)), -cos(radians(theta))]])
	
	shift_p = qutip.Qobj(np.roll(np.eye(P), 1, axis=0))
	shift_m = qutip.Qobj(np.roll(np.eye(P),-1, axis=0))
	shift = tensor(ket0*ket0.dag(),shift_p) + tensor(ket1*ket1.dag(),shift_m)	
	
	U = shift*(tensor(coin,qeye(P)))
	return U

#Non-Unitary quantum walk:
def Qwalk_gen(step_size,qubit_state,theta,gamma,nu):
	P=2*step_size+1
	P_state = basis(P,step_size)						
	Psi = ket2dm(tensor(qubit_state,P_state))		
	W = Walk(theta,step_size) 

	for t in range(step_size):	
		if t != 0:
			tau = t*nu; delta = sqrt(gamma**2-4)
			G = exp(-gamma*tau)*((np.cosh(delta*tau)+gamma*np.sinh(delta*tau)/delta))
			K1 = tensor(sqrt((1+G)/2.0)*qeye(2), qeye(P))
			K2 = tensor(sqrt((1-G)/2.0)*sigmaz(),qeye(P))
			Psi = W*(K1*Psi*K1.dag()+K2*Psi*K2.dag())*W.dag()
	return Psi

#variance as a function of time
def Variance_rate(gamma):
	var=[0];t=[0]
	step_size=51;qubit_state=psip;theta=45;nu=0.1
	for step in range(1,step_size):
	    qw = Qwalk_gen(step,qubit_state,theta,gamma,nu);v=[]
	    sites=2*step+1
	    site = [-step+i for i in range(0,sites)]
	    t.append(step)
	    for i in site:
	        v.append(qw[i]*pow(site[i]-np.average(qw),2))
	    var.append(sum(v))
	return var

#Instantaneous variance
def Walk_variance(gamma):
	var=[0];t=[0];step_size=51;qubit_state=psip;theta=45;nu=0.1
	qw = Qwalk_gen(step_size,qubit_state,theta,gamma,nu);v=[]
	sites=2*step_size+1
	site = [-step_size+i for i in range(0,sites)]
	t.append(step_size)
	for i in site:
		v.append(qw[i]*pow(site[i]-np.average(qw),2))
	return sum(v) 

#Von Neumann entropy
def von_entropy(state):
    ev = state.eigenenergies()
    Vn_ent=[]
    for i in range(len(ev)):
        Vn_ent.append(-ev[i]*np.log2(ev[i])) 
    Vn_ent = filter(lambda v: v==v, Vn_ent) #to remove nans from the list
    return sum(Vn_ent)

#Mutual Information
def mutual_info(rho):
    rhoA = rho.ptrace(0)
    rhoB = rho.ptrace(1)
    out =  von_entropy(rhoA)+von_entropy(rhoB)-von_entropy(rho)
    return out

#Spectral decomposition
def projectors(qstate):
    ev, evec = qstate.eigenstates();
    m=[]
    for l in range(len(evec)):
        m.append(ket2dm(evec[l]))
    return m

#Measurement Induced disturbance
def quantum_mid(walk_steps,freq,amp):
    proj_w=[];proj_rho=[]
    rho = Qwalk_gen(walk_steps,psip,45,freq,amp)
    rho_c = rho.ptrace(0); rho_p = rho.ptrace(1)
    proj_c = projectors(rho_c); proj_p = projectors(rho_p)
    for n in range(len(proj_c)):
        for p in range(len(proj_p)):
            M = tensor(proj_c[n],proj_p[p])
            proj_rho.append(M*rho*M.dag())
    rhoMID = sum(proj_rho)
    Im_rho = mutual_info(rho)
    Im_mid = mutual_info(rhoMID)
    return Im_rho-Im_mid

#Quantum discord:
def quantum_discord(t):
    rho_pc=Qwalk_gen(t,psip,45,0.001,0.05)
    sites = 2*t+1
    Discord=[]
    for teta in range(0,180,9):
        for phi in range(0,360,13):
            E = qubit_gen(teta,phi).eigenstates()
            Eval = E[0];Ev = E[1]; Evec0=ket2dm(Ev[0]); Evec1=ket2dm(Ev[1])
            
            Ipc = von_entropy(rho_pc.ptrace(0))-von_entropy(rho_pc)
            p0=(tensor(Evec0,qeye(sites))*rho_pc).tr(); p1=(tensor(Evec1,qeye(sites))*rho_pc).tr()
            
            rho_c1 = von_entropy((tensor(Evec0,qeye(sites))*rho_pc*tensor(Evec0,qeye(sites)))/p0)
            rho_c2 = von_entropy((tensor(Evec1,qeye(sites))*rho_pc*tensor(Evec1,qeye(sites)))/p1)
            z = Ipc+(p0*rho_c1+p1*rho_c2)
            Discord.append(z.real)
    return min(Discord)

#Trace distance (Neilsen & Chuang)
def TraceD(rho1,rho2):
    diff = rho1 - rho2
    Eval = diff.eigenenergies()
    return 0.5*sum(abs(Eval))

#Bloch sphere sampling
def qubit_gen(teta,phi):
    psi = cos(radians(teta))*ket0+complex(cos(radians(phi)),sin(radians(phi)))*sin(radians(teta))*ket1
    return ket2dm(psi.unit())   
```

```python
#Quantum discord
D=[0]
L=[p for p in range(1,102)]
if __name__ == '__main__':
    p = Pool()
    D = p.map(quantum_discord, [t for t in L])
    p.close()
    p.join()
```
