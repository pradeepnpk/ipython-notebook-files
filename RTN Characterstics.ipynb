```python
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
from math import *
from qutip import *
from cmath import pi, sin, cos, sqrt,exp; from math import radians
import seaborn as sns

#-------------------------EXPONENTIAL AUTO-CORRELATION-------------------------#
def auto_correlation(gamma,t):
    amp=1.0
    kai = (amp**2)*np.exp(-2*gamma*t)
    return kai

#-------------------------LORRENZIAN SPECTRUM----------------------------------#
def lorrenzian(gamma,f):
    RTN_spectrum = (4*gamma)/(4*(pi**2)*(f**2)+gamma**2)
    return RTN_spectrum

#-------------------------KERNAL-----------------------------------------------#
def kernal(gamma,t):
    amp=0.1;tau=1/(2*gamma)
    nu = t*gamma; mu = sqrt((2.0*amp*(1/gamma))**2-1.0)
    gamma = np.exp(-nu)*(np.cos(nu*mu)+np.sin(nu*mu)/mu)#np.cos handles complex argument as cosh. Alternatively, np.cosh can be used but im(mu) has to be obtained separately      
    return abs(gamma)

#------------------------Fluctuation parameter--------------------------------#
def fluctuation(tau):
    return 1/(2.0*tau)
```

```python
f = np.linspace(-1,1,500)
spectrum1 = lorrenzian(0.05,f)
spectrum2 = lorrenzian(0.10,f)
spectrum3 = lorrenzian(0.50,f)
spectrum4 = lorrenzian(1.0,f)

tac = np.linspace(0,1.0,100)
ac1 = auto_correlation(0.05,tac)
ac2 = auto_correlation(0.10,tac)
ac3 = auto_correlation(0.50,tac)
ac4 = auto_correlation(1.00,tac)

t = np.linspace(0,100,101)
k1 = kernal(0.05,t)
k2 = kernal(0.10,t)
k3 = kernal(0.50,t)
k4 = kernal(1.00,t)

tau = np.linspace(0,1,101)
f1 = fluctuation(tau)
```

```python
fig = plt.figure(figsize=(8,6))
sns.set_context('notebook',font_scale=1.5)
sns.set_style('ticks')
ax = fig.add_subplot(221)
plt.title(r'$Lorrentzian \ power \ spectrum$')
plt.xlabel(r'$\omega$');plt.ylabel(r'$S_{\eta}(\omega)$')
plt.plot(f,spectrum1,label=r'$\gamma=0.05$');plt.plot(f,spectrum2,label=r'$\gamma=0.10$')
plt.plot(f,spectrum3,label=r'$\gamma=0.50$');plt.plot(f,spectrum4,label=r'$\gamma=1.0$')
legend = ax.legend(loc='upper right');

ax = fig.add_subplot(222)
plt.title(r'$Autocorrelation$')
plt.xlabel(r'$t$');plt.ylabel(r'$\eta(t)$')
plt.plot(tac,ac1,label=r'$\gamma=0.05$');plt.plot(tac,ac2,label=r'$\gamma=0.10$')
plt.plot(tac,ac3,label=r'$\gamma=0.50$');plt.plot(tac,ac4,label=r'$\gamma=1.0$')
legend = ax.legend(loc='upper right')

ax = fig.add_subplot(223)
plt.title(r'$Memory \ kernal$')
plt.xlabel(r'$t$');plt.ylabel(r'$\Lambda(t)$')
plt.plot(t,k1,label=r'$\gamma=0.05$');plt.plot(t,k2,label=r'$\gamma=0.10$')
plt.plot(t,k3,label=r'$\gamma=0.50$');plt.plot(t,k4,label=r'$\gamma=1.0$')
legend = ax.legend(loc='upper right')

ax = fig.add_subplot(224)
plt.title(r'$\gamma \ vs \ \tau $')
plt.xlabel(r'$\tau$');plt.ylabel(r'$\gamma = \tau^{-1}$')
plt.plot(t,f1,label=r'$\tau$')
legend = ax.legend(loc='upper right')
sns.despine()
plt.show()
```
