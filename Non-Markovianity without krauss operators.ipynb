```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt
from cmath import pi, sin, cos, sqrt,exp; from math import radians
import seaborn as sns;	sns.set_style('ticks')
from multiprocessing import Pool
from scipy import linalg; from scipy import misc
from scipy.signal import argrelextrema
import time as time
from random import randrange

#Basis states:
ket0 = ket2dm(basis(2,0).unit())
ket1 = ket2dm(basis(2,1).unit())
psip = ket2dm((basis(2,0)+basis(2,1)*1j).unit())
psim = ket2dm((basis(2,0)-basis(2,1)*1j).unit())
psiplus = ket2dm((basis(2,0)+basis(2,1)).unit())
psiminus = ket2dm((basis(2,0)-basis(2,1)).unit())

#Walk operator:
def Walk(theta,step_size):
    P = 2*step_size+1
    coin = qutip.Qobj([[cos(radians(theta)), sin(radians(theta))],
                       [sin(radians(theta)), -cos(radians(theta))]])
    
    shift_p = qutip.Qobj(np.roll(np.eye(P), 1, axis=0))
    shift_m = qutip.Qobj(np.roll(np.eye(P),-1, axis=0))
    shift = tensor(ket0*ket0.dag(),shift_p) + tensor(ket1*ket1.dag(),shift_m)   
    
    U = shift*(tensor(coin,qeye(P)))
    return U

#Non-Unitary quantum walk:
def Qwalk_gen(step_size,qubit_state,theta,gamma,nu):
    P=2*step_size+1
    P_state = ket2dm(basis(P,step_size))                        
    Psi = tensor(qubit_state,P_state)     
    W = Walk(theta,step_size) 

    for t in range(step_size):  
        if t != 0:
            tau = t*nu; delta = sqrt(gamma**2-4)
            G = exp(-gamma*tau)*((np.cosh(delta*tau)+gamma*np.sinh(delta*tau)/delta))
            K1 = tensor(sqrt((1+G)/2.0)*qeye(2), qeye(P))
            K2 = tensor(sqrt((1-G)/2.0)*sigmaz(),qeye(P))
            Psi = W*(K1*Psi*K1.dag()+K2*Psi*K2.dag())*W.dag()
#     #MEASUREMENT
#     Prob=[];step=[]
#     for k in range(0,P,1):
#         Proj = basis(P,k)*basis(P,k).dag()
#         Measure = tensor(qeye(2),Proj)
#         q = abs((Psi*Measure).tr())
#         Prob.append(q)
#         step.append(k)
    return Psi

#variance as a function of time
def Variance_rate(gamma):
    var=[0];t=[0]
    step_size=51;qubit_state=psip;theta=45;nu=0.08
    for step in range(1,step_size):
        qw = Qwalk_gen(step,qubit_state,theta,gamma,nu);v=[]
        sites=2*step+1
        site = [-step+i for i in range(0,sites)]
        t.append(step)
        for i in site:
            v.append(qw[i]*pow(site[i]-np.average(qw),2))
        var.append(sum(v))
    return var

#Instantaneous variance
def Walk_variance(nu):
    v=[0]
    step_size = 101;qubit_state=psip;coin_angle=45; gamma=0.001
    qw = Qwalk_gen(step_size,qubit_state,coin_angle,gamma,nu);v=[]
    sites=2*step_size+1
    site = [-step_size+i for i in range(0,sites)]
    for i in site:
        v.append(qw[i]*pow(site[i]-np.average(qw),2))
    return sum(v)

def von_entropy(state):
    ev = state.eigenenergies()
    Vn_ent=[]
    for i in range(len(ev)):
        Vn_ent.append(-ev[i]*np.log2(ev[i])) 
    Vn_ent = filter(lambda v: v==v, Vn_ent) #to remove nans from the list
    return sum(Vn_ent)

def mutual_info(rho):
    rhoA = rho.ptrace(0)
    rhoB = rho.ptrace(1)
    out =  von_entropy(rhoA)+von_entropy(rhoB)-von_entropy(rho)
    return out

def projectors(qstate):
    ev, evec = qstate.eigenstates()
    m=[]
    for l in range(len(evec)):
        m.append(ket2dm(evec[l]))
    return m

#trace distance measure
def Qwalk_traced(rho1,rho2,gamma,nu): 
    Ntd = [TraceD(rho1,rho2)] #density matrices needed
    for t in range(1,51):
        Nrho1 = Qwalk_gen(t,rho1,45,gamma,nu).ptrace(0) #reduced psip
        Nrho2 = Qwalk_gen(t,rho2,45,gamma,nu).ptrace(0) #reduced psim
        Ntd.append(TraceD(Nrho1,Nrho2))
    return Ntd

#Trace distance (Neilsen & Chuang)
#1/2*tr.[sqrt((rho1-rho2)^dag (rho1-rho2))]
def TraceD(rho1,rho2):
    diff = rho1 - rho2
    Eval = diff.eigenenergies()
    return 0.5*sum(abs(Eval))
```

```python
#PLOT OF TRACE DISTANCE FOR THE OPTIMAL QUBITS
Tdistance_nonoise = Qwalk_traced(psiplus,psiminus,0.0,0.01)
plt.plot(Tdistance_nonoise,label='Unitary')
plt.show()
```

```python
#-----------------------------------RATE OF CHANGE OF TRACE DISTANCE: SIGMA-------------------------------------#
def blp_measure(gamma):
    blp=[];nu=0.1
    def sigma(t):
        dist=0
        if t != 0:
            rho1 = Qwalk_gen(int(t),psiplus,45,gamma,nu).ptrace(0)
            rho2 = Qwalk_gen(int(t),psiminus,45,gamma,nu).ptrace(0)
            dist = TraceD(rho1,rho2) # D(\rho1(t),\rho2(t))
        return dist
        
    nmarkov_diff=[]
    for y in range(1,51):
        nmarkov_diff.append(misc.derivative(sigma,y))
    
    measure=[]    
    for i in range(0,len(nmarkov_diff)):
        if nmarkov_diff[i]>0:
            measure.append(nmarkov_diff[i])
    blp.append(np.trapz(measure)) 
    return blp

# PARALLEL PROCESSING 
F=[x for x in np.arange(0.01,3.01,0.1)]
if __name__ == '__main__':
    p = Pool()
    N_blp = p.map(blp_measure, [gamma for gamma in F])
    p.close()
    p.join()
```
