```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt
from cmath import pi, sin, cos, sqrt,exp
from math import radians
import seaborn as sns
import time
from multiprocessing import Pool
from random import randrange

#Basis states
ket0 = basis(2,0)
ket1 = basis(2,1)
psip = (basis(2,0)+basis(2,1)*1j).unit()
psim = (basis(2,0)-basis(2,1)*1j).unit()
psiplus = (basis(2,0)+basis(2,1)).unit() 
psiminus = (basis(2,0)-basis(2,1)).unit()


#Walk operator
def Walk_op(step_size,coin_angle):
	sites = 2*step_size+1
	coin = qutip.Qobj([[cos(radians(coin_angle)), sin(radians(coin_angle))],
	   [sin(radians(coin_angle)), -cos(radians(coin_angle))]])

	shift_p = qutip.Qobj(np.roll(np.eye(sites), 1, axis=0))
	shift_m = qutip.Qobj(np.roll(np.eye(sites),-1, axis=0))
	shift = tensor(ket0*ket0.dag(),shift_p) + tensor(ket1*ket1.dag(),shift_m)	

	W = shift*(tensor(coin,qeye(sites)))
	return W

#Quantum walk generator
def Qwalk_gen(step_size,qubit_state,coin_angle,freq,amp):
	sites=2*step_size+1
	P_state = basis(sites,step_size)						
	Psi = ket2dm(tensor(qubit_state,P_state))		
	W = Walk_op(step_size,coin_angle) 

	for t in range(step_size):	
		nu = t*freq; mu = sqrt((2.0*amp*(1/freq))**2-1.0)
		gamma = exp(-nu)*(np.cos(nu*mu)+np.sin(nu*mu)/mu)
		K1 = tensor(sqrt((1+gamma)/2.0)*qeye(2), qeye(sites))
		K2 = tensor(sqrt((1-gamma)/2.0)*sigmaz(),qeye(sites))
		Psi = W*(K1*Psi*K1.dag()+K2*Psi*K2.dag())*W.dag()
	return Psi

#Uhlmann's fidelity for mixed sattes (Jozsa's paper)
def quantum_fidelity(rho1, rho2):
	rho_ax = (rho1.sqrtm()*rho2*rho1.sqrtm()).sqrtm()
	return (rho_ax.tr())**2

#von nemann entropy
def von_entropy(state):
    ev = state.eigenenergies()
    Vnent=[]
    for i in range(len(ev)):
        Vnent.append(-ev[i]*np.log2(ev[i])) 
    Vnent = filter(lambda v: v==v, Vnent) #to remove nans from the list
    return sum(Vnent)

#quantum mutual information (Wiki)
def mutual_info(rho):
    rhoA = rho.ptrace(0)
    rhoB = rho.ptrace(1)
    out =  von_entropy(rhoA)+von_entropy(rhoB)-von_entropy(rho)
    return out

# if __name__ == "__main__":

# 	#F(\rho_0,\rho_tau)
# 	fid_markov_0=[]; fid_nonmarkov_0=[]
# 	rho_0 = Qwalk_gen(0,psip,45,0.001,0.1).ptrace(0)
# 	for tau in range(0,101):		
# 		rho_tau_markov = Qwalk_gen(tau,psip,45,7.0,0.1).ptrace(0)
# 		rho_tau_nonmarkov = Qwalk_gen(tau,psip,45,0.001,0.1).ptrace(0)
# 		fid_markov_0.append(quantum_fidelity(rho_0,rho_tau_markov))
# 		fid_nonmarkov_0.append(quantum_fidelity(rho_0,rho_tau_nonmarkov))
	
# 	fig = plt.figure(figsize=(8,6))
# 	sns.set_context('notebook',font_scale=2.3)
# 	sns.set_style('ticks')
# 	ax = fig.add_subplot(111)
# 	plt.plot(fid_markov_0,label='Markovian')
# 	plt.plot(fid_nonmarkov_0,label='Non-Markovian')
# 	legend = ax.legend(loc='upper right');
# 	plt.show()

# 	#F(\rho(t),\rho(t+\tau))
# 	fid_markov_0=[]; fid_nonmarkov_0=[]
# 	tau=1
# 	for t in range(0,101):	
# 		rho_markov_0 = Qwalk_gen(t,psip,45,7.0,0.01).ptrace(0)
# 		rho_tau_markov = Qwalk_gen(t+tau,psip,45,7.0,0.01).ptrace(0)
		
# 		rho_nonmarkov_0 = Qwalk_gen(t,psip,45,0.001,0.01).ptrace(0)
# 		rho_tau_nonmarkov = Qwalk_gen(t+tau,psip,45,0.001,0.01).ptrace(0)
		
# 		fid_markov_0.append(quantum_fidelity(rho_markov_0,rho_tau_markov))
# 		fid_nonmarkov_0.append(quantum_fidelity(rho_nonmarkov_0,rho_tau_nonmarkov))

# 	fig = plt.figure(figsize=(8,6))
# 	sns.set_context('notebook',font_scale=2.3)
# 	sns.set_style('ticks')
# 	ax = fig.add_subplot(111)
# 	plt.plot(fid_markov_0,label='Markovian')
# 	plt.plot(fid_nonmarkov_0,label='Non-Markovian')
# 	legend = ax.legend(loc='upper right');
# 	plt.show()
```

```python
#density matrix from the paper

def qstate_AB(t,tau):
    gamma=0.0001; Lambda=0.5;
    rhoAB_t = qutip.Qobj([[1,0,0,0],
                          [0,4,Lambda*exp(-(gamma*t)/2),0],
                          [0,Lambda*exp(-(gamma*t)/2),4,0],
                          [0,0,0,0]])
    
    rhoAB_tau = qutip.Qobj([[1,0,0,0],
                          [0,4,Lambda*exp(-(gamma*t+tau)/2),0],
                          [0,Lambda*exp(-(gamma*t+tau)/2),4,0],
                          [0,0,0,0]])
    
    fid = fidelity((rhoAB_t).ptrace(0),(rhoAB_tau).ptrace(0)) 
    return fid
```

```python
q=[]
qinit = qstate_AB(0,1) #F(\rho(0),\rho(\tau))
for t in np.arange(0,1,0.001):
    qfinal = qstate_AB(t,1) #F(\rho(t),\rho(t+\tau))
    q.append((qfinal-qinit)/qinit)
```

```python
plt.plot(q);plt.show()
```
