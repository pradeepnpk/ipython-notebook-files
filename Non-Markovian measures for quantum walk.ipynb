```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt
from cmath import pi, sin, cos, sqrt,exp; from math import radians
import seaborn as sns;	sns.set_style('ticks')
from multiprocessing import Pool
from scipy import linalg; from scipy import misc
from scipy.signal import argrelextrema
import time as time
from random import randrange

#Basis states:
ket0 = ket2dm(basis(2,0).unit())
ket1 = ket2dm(basis(2,1).unit())
psip = ket2dm((basis(2,0)+basis(2,1)*1j).unit())
psim = ket2dm((basis(2,0)-basis(2,1)*1j).unit())
psiplus = ket2dm((basis(2,0)+basis(2,1)).unit())
psiminus = ket2dm((basis(2,0)-basis(2,1)).unit())

#Walk operator:
def Walk(coin_angle,step_size):
	P = 2*step_size+1
	coin = qutip.Qobj([[cos(radians(coin_angle)), sin(radians(coin_angle))],
				       [sin(radians(coin_angle)), -cos(radians(coin_angle))]])
	
	shift_p = qutip.Qobj(np.roll(np.eye(P), 1, axis=0))
	shift_m = qutip.Qobj(np.roll(np.eye(P),-1, axis=0))
	shift = tensor(ket0*ket0.dag(),shift_p) + tensor(ket1*ket1.dag(),shift_m)	
	
	U = shift*(tensor(coin,qeye(P)))
	return U

def Coin(coin_angle):
    coin = qutip.Qobj([[cos(radians(coin_angle)), sin(radians(coin_angle))],
                       [sin(radians(coin_angle)), -cos(radians(coin_angle))]])
    return coin


def Qwalk_gen(step_size,qubit_state,coin_angle,freq,amp):
    P=2*step_size+1
    P_state = ket2dm(basis(P,step_size))						
    Psi = tensor(qubit_state,P_state)
    W = Walk(coin_angle,step_size) 

    for t in range(step_size):	
        nu = t*freq; mu = sqrt((2.0*amp*(1/freq))**2-1.0)
        gamma = exp(-nu)*(np.cos(nu*mu)+np.sin(nu*mu)/mu)
        K1 = tensor(sqrt((1+gamma)/2.0)*qeye(2), qeye(P))
        K2 = tensor(sqrt((1-gamma)/2.0)*sigmaz(),qeye(P))
        Psi = W*(K1*Psi*K1.dag()+K2*Psi*K2.dag())*W.dag()      
    return Psi

def Qwalk_gen_markov(step_size,qubit_state,coin_angle,p):
    sites=2*step_size+1
    P_state = ket2dm(basis(sites,step_size))           
    Psi = tensor(qubit_state,P_state) 
    W = Walk(coin_angle,step_size)           
    
    for i in range(step_size):              
        Psi = W*GAmpdamping(step_size,Psi,p)*W.dag()   
    return Psi

#Trace distance (Neilsen & Chuang)
#1/2*tr.[sqrt((rho1-rho2)^dag (rho1-rho2))]
def TraceD(rho1,rho2):
    diff = rho1 - rho2
    Eval = diff.eigenenergies()
    return 0.5*sum(abs(Eval))

def Qfidelity(rho1, rho2):
    rho_ax = (rho1.sqrtm()*rho2*rho1.sqrtm()).sqrtm()
    return ((rho_ax.tr())**2).real

#Bloch sphere sampling
def Qubit_gen(teta,phi):
    psi = cos(radians(teta))*ket0+complex(cos(radians(phi)),sin(radians(phi)))*sin(radians(teta))*ket1
    return ket2dm(psi.unit()) 

#Generates a random qubit:
def Rand_qubit():
    theta = randrange(0,180,10)
    phi = randrange(0,360,10)
    return Qubit_gen(theta,phi)

#projection matrix 
def Projection_op(qstate):
    ev, evec = qstate.eigenstates()
    m=[]
    for l in range(len(evec)):
        m.append(ket2dm(evec[l]))
    return m

def Dephasing(qstate,p):
    dstate = (1-p)*qstate+p*(tensor(sigmaz(),qeye(P))*qstate*tensor(sigmaz(),qeye(P)))
    return dstate

#DEPOLARIZING CHANNEL
def Depolarizing(qstate,p):
    dstate = (1-p)*qstate+p/3*(tensor(sigmax(),qeye(P))*(qstate)*tensor(sigmax(),qeye(P))*
                               tensor(sigmay(),qeye(P))*(qstate)*tensor(sigmay(),qeye(P))*
                               tensor(sigmaz(),qeye(P))*(qstate)*tensor(sigmaz(),qeye(P)))
    return dstate

#AMPLITUDE DAMPING
def Ampdampling(qstate,p):
    M1 = qutip.Qobj([[1,        0],  # M-> KRAUSS OPERATORS
                     [0,sqrt(1-p)]])

    M2 = qutip.Qobj([[0,  sqrt(p)],
                     [0,        0]])
    dstate = tensor(M1,qeye(P))*(qstate)*tensor(M1.dag(),qeye(P))+tensor(M2,qeye(P))*(qstate)*tensor(M2.dag(),qeye(P))
    return dstate 

#GENERALIZED AMPLITUDE DAMPING
def GAmpdamping(step_size,qstate,p):
    kai = 1; sites = 2*step_size+1
    M1 = qutip.Qobj([[sqrt(kai),     0              ],
                     [0,        sqrt(kai)*sqrt(1-p) ]])

    M2 = qutip.Qobj([[0,        sqrt(kai)*sqrt(p)   ],
                     [0,        0                   ]])
    
    M3 = qutip.Qobj([[sqrt(1-kai)*sqrt(1-p),       0],
                     [0,                 sqrt(1-kai)]])

    M4 = qutip.Qobj([[0,                           0],
                     [sqrt(1-kai)*sqrt(p),         0]])
    
    dstate = (tensor(M1,qeye(sites))*(qstate)*tensor(M1.dag(),qeye(sites))+
              tensor(M2,qeye(sites))*(qstate)*tensor(M2.dag(),qeye(sites))+
              tensor(M3,qeye(sites))*(qstate)*tensor(M3.dag(),qeye(sites))+
              tensor(M4,qeye(sites))*(qstate)*tensor(M4.dag(),qeye(sites)))
    return dstate

def von_entropy(state):
    ev = state.eigenenergies()
    Vn_ent=[]
    for i in range(len(ev)):
        Vn_ent.append(-ev[i]*np.log2(ev[i])) 
    Vn_ent = filter(lambda v: v==v, Vn_ent) #to remove nans from the list
    return sum(Vn_ent)

def mutual_info(rho):
    rhoA = rho.ptrace(0)
    rhoB = rho.ptrace(1)
    out =  von_entropy(rhoA)+von_entropy(rhoB)-von_entropy(rho)
    return out

def projectors(qstate):
    ev, evec = qstate.eigenstates()
    m=[]
    for l in range(len(evec)):
        m.append(ket2dm(evec[l]))
    return m

#trace distance measure
def Qwalk_traced(rho1,rho2,freq,amp): 
    Ntd = [TraceD(rho1,rho2)] #density matrices needed
    for t in range(1,51):
        Nrho1 = Qwalk_gen(t,rho1,45,freq,amp).ptrace(0) #reduced psip
        Nrho2 = Qwalk_gen(t,rho2,45,freq,amp).ptrace(0) #reduced psim
        Ntd.append(TraceD(Nrho1,Nrho2))
    return Ntd

#BLP Optimization with random qubits
def Blp_optimization(freq,amp):
    Blpmax=0;numofqbits=100
    for op in range(0,numofqbits): #100 random qubits 
        rho1, rho2 = Projection_op(Rand_qubit())
        td = Qwalk_traced(rho1,rho2,freq,amp)
        Blpm =  Blp_measure(td)
        if Blpm > Blpmax:
            Blpmax = Blpm
    return Blpmax
```

```python
#-----------------------------------RATE OF CHANGE OF TRACE DISTANCE: SIGMA-------------------------------------#
def blp_measure(f):
    blp=[];amp=0.8
    def sigma(t):
        dist=0
        if t != 0:
            rho1 = Qwalk_gen(int(t),psiplus,45,f,amp).ptrace(0)
            rho2 = Qwalk_gen(int(t),psiminus,45,f,amp).ptrace(0)
            dist = TraceD(rho1,rho2) # D(\rho1(t),\rho2(t))
        return dist
        
    nmarkov_diff=[]
    for y in range(1,51):
        nmarkov_diff.append(misc.derivative(sigma,y))
    
    measure=[]    
    for i in range(0,len(nmarkov_diff)):
        if nmarkov_diff[i]>0:
            measure.append(nmarkov_diff[i])
    blp.append(np.trapz(measure)) 
    return blp

# PARALLEL PROCESSING 
F=[x for x in np.arange(0.001,7.0,0.01)]
if __name__ == '__main__':
    p = Pool()
    N_blp = p.map(blp_measure, [freq for freq in F])
    p.close()
    p.join()
```

```python
#COMPUTING CHANNEL CAPACITY USING ANALYTICAL EQUATION: 1 - H_2(1-gamma(nu)/2)
def Quantum_channal_capacity(t,freq,amp):
    nu = t*freq; mu = sqrt((2.0*amp*(1/freq))**2-1.0)
    gamma = exp(-nu)*(np.cos(nu*mu)+np.sin(nu*mu)/mu)
    p = (1-gamma)/2
    binary_ent = -p*np.log2(p)-(1-p)*np.log2(1-p)
    Qcc = 1-binary_ent
    return Qcc

Qcapacity_m = [1]; Qcapacity_nm = [1]
for t in np.arange(1,101,0.1):
    Qcapacity_m.append(Quantum_channal_capacity(t,0.001,0.25)) 
    Qcapacity_nm.append(Quantum_channal_capacity(t,7.0,0.25)) 
    
x = [x for x in np.arange(0,100)]
plt.plot(Qcapacity_m,lw=2)
plt.plot(Qcapacity_nm,lw=2)
plt.show()
```
