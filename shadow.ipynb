```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from math import sin, asin, cos,acos,tan, pi, radians, degrees
%matplotlib inline

def hour_angle(time):
    noon=12
    hAout = 15*(time-noon)
    return hAout #in degrees

def declination_angle(daynum):
    dAout = radians(23.45)*sin(radians(360*(284+daynum)/365))
    return degrees(dAout) #in degrees

def elevation_angle(time,daynum):
    lattitude = radians(26.3) #Jodhpur, India
    hA = radians(hour_angle(time))
    dA = radians(declination_angle(daynum))
    eAout = np.arcsin(cos(lattitude)*cos(dA)*cos(hA)+sin(lattitude)*sin(dA))
    return degrees(eAout)

def zenith_angle(time,daynum):
    eA = elevation_angle(time,daynum)
    zA = 90-eA
    return zA

def Azimuth_angle(time,daynum):
    lattitude=radians(26.3)
    zA = radians(zenith_angle(time,daynum))
    dA = radians(declination_angle(daynum))
    aA = np.arccos(sin(zA)*sin(lattitude)-sin(dA)/cos(zA)*cos(lattitude))
    return degrees(aA)

def Shadow_length(time,daynum,pole_height):
    eA = elevation_angle(time,daynum)
    shadowLen = pole_height/np.tan(radians(eA))
    return shadowLen
```

```python
#------------------------------------> SOLAR ANGLES TEST <--------------------------------------------------------#
eAP=[]
for d in range(1,366):
    eAP.append(elevation_angle(9,d))
plt.plot(eAP,lw=3)
plt.xlim([1,365])
plt.show()
```

```python
#----------------------------------> SHADOW LENGTH CALC <---------------------------------------------------------#
time = 12 #in 24hrs
pole_height = 5 # in meters
Slength=[]
for sl in range(1,366):
    Slength.append(Shadow_length(time,sl,pole_height))
plt.plot(Slength,lw=3);plt.show()
```
